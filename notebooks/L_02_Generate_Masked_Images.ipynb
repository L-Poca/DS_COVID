```python
# importation des bibliothèques basiques
from pathlib import Path
import sys
import os
from IPython.display import display

# Ajout du répertoire racine du projet au PYTHONPATH pour tous les imports
project_root = Path().resolve().parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

# Ensuite importer les modules nécessaires
import src.features.Traitement_Data.traitement_modulaire as traitement_modulaire
import src.features.Traitement_Data.C_traitement as C_traitement


# Configuration de l'affichage des graphiques
paths = C_traitement.Get_All_Paths_2()
dsize = (256, 256)

# Fonction pour afficher les images et leurs statistiques
import matplotlib.pyplot as plt
def afficher_images_et_stats(images_masked, stats_list, FileName):
    """Affiche les images et leurs statistiques."""

    # Affichage des images et statistiques
    print(f"Displaying first 5 images and stats for {FileName}...")
    for i, img in enumerate(images_masked[:5]):
        plt.figure()
        cmap='gray' if len(img.shape)==2 else "RGB"
        plt.imshow(img, cmap=cmap)
        plt.title(f"Image {img.shape} - {FileName} - {cmap}")
        plt.axis('off')
        plt.show()
        display(stats_list[i])


# Traitement des images par catégorie
print("Starting image processing...")
for key, path_dict in paths.items():

    # Récupération des chemins d'image et de masque
    img_path = path_dict.get("img")
    mask_path = path_dict.get("mask")

    # Vérification de l'existence des masques
    print(f"Processing images in {key} {'with masks' if mask_path else 'without masks'}...")
    
    # Nom du fichier de sortie
    FileName = f"{key}_{dsize[0]}x{dsize[1]}{'_masked' if mask_path else ''}"

    # Traitement des images
    save_images = True
    df, images_masked, stats_list = traitement_modulaire.main_workflow(
        image_folder=img_path,
        dsize=dsize,
        Categorie= key,
        grayscale=True,
        flat_array=True,
        mask_folder=mask_path,
        show_each=False,
        FileName=FileName,
        export_csv=False,
        save_images=True # pour sauvegarder les images traitées, Dossier-Automatique par resolution et categories
    )
    # Affichage des résultats
    #display(df.head())
    afficher_images_et_stats(images_masked, stats_list, FileName)
```
