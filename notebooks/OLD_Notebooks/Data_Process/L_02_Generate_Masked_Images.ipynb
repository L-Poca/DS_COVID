```python
# importation des bibliothèques basiques
from pathlib import Path
import sys
import os
from IPython.display import display
import pandas as pd

# Ajout du répertoire racine du projet au PYTHONPATH pour tous les imports
project_root = Path().resolve().parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

# Ensuite importer les modules nécessaires
import src.features.Traitement_Data.traitement_modulaire as traitement_modulaire
import src.features.Traitement_Data.C_traitement as C_traitement


# Configuration de l'affichage des graphiques
paths = C_traitement.Get_All_Paths_2()
dsize = (256, 256)  # Taille de redimensionnement des images

# Fonction pour afficher les images et leurs statistiques
import matplotlib.pyplot as plt
def afficher_images_et_stats(images_masked, stats_list, FileName):
    """Affiche les images et leurs statistiques."""

    # Affichage des images et statistiques
    print(f"Displaying first 5 images and stats for {FileName}...")
    for i, img in enumerate(images_masked[:5]):
        plt.figure()
        cmap='gray' if len(img.shape)==2 else "RGB"
        plt.imshow(img, cmap=cmap)
        plt.title(f"Image {img.shape} - {FileName} - {cmap}")
        plt.axis('off')
        plt.show()
        display(stats_list[i])

def Afficher_statistiques(list_stat, FileName):
    """Affiche les statistiques du DataFrame."""
    print(f"Displaying statistics for {FileName}...")
    # Création d'un DataFrame à partir de la liste de statistiques
    df = pd.DataFrame(list_stat)
    df['FileName'] = FileName  # Ajout du nom de fichier pour référence
    # Affichage des statistiques
    display(df)
    display(df.describe())
    display(df.head())

    #visualisation des statistiques
    df.hist(figsize=(10, 10), bins=30)
    # Affichage des graphiques
    plt.suptitle(f"Statistiques pour {FileName}")
    plt.show()







# Traitement des images par catégorie
print("Starting image processing...")

Stats = {}

for key, path_dict in paths.items():

    # Récupération des chemins d'image et de masque
    img_path = path_dict.get("img")
    mask_path = path_dict.get("mask")

    # Vérification de l'existence des masques
    print(f"Processing images in {key} {'with masks' if mask_path else 'without masks'}...")
    
    # Nom du fichier de sortie
    FileName = f"{key}_{dsize[0]}x{dsize[1]}{'_masked' if mask_path else ''}"

    # Traitement des images
    save_images = True
    df, images_masked, stats_list = traitement_modulaire.main_workflow(
        image_folder=img_path,
        dsize=dsize,
        Categorie= key,
        grayscale=True,
        flat_array=True,
        mask_folder=mask_path,
        show_each=False,
        FileName=FileName,
        export_csv=False, # ne pas utiliser pour l'instant, c'est les numpy arrays des images qui sont sauvegardés
        save_images=True # pour sauvegarder les images traitées, Dossier-Automatique par resolution et categories
    )

    # Sauvegarde des statistiques dans le dictionnaire
    Stats[key] = stats_list

    # Affichage des résultats
    #display(df.head())
    #afficher_images_et_stats(images_masked, stats_list, FileName)

for key, stats_list in Stats.items():
    # Affichage des statistiques pour chaque catégorie
    Afficher_statistiques(stats_list, FileName)
```

```python
# Comparaison des valeurs de chaque catégorie par métrique avec un diagramme chart (boxplot)
import matplotlib.pyplot as plt
import seaborn as sns

# Reconstituer les DataFrames de stats pour chaque catégorie à partir de Stats
all_stats = []
for key, stats_list in Stats.items():
    df_stats = pd.DataFrame(stats_list)
    df_stats['Categorie'] = key
    all_stats.append(df_stats)

# Concaténer tous les DataFrames
df_all = pd.concat(all_stats, ignore_index=True)

# Sélectionner les colonnes numériques à comparer (hors FileName/Categorie)
metrics = [col for col in df_all.columns if col not in ['FileName', 'Categorie']]

# Afficher un diagramme en boîte (boxplot) pour chaque métrique par catégorie
plt.figure(figsize=(16,10))
for i, metric in enumerate(metrics):
    plt.subplot(1, len(metrics), i+1)
    sns.boxplot(x='Categorie', y=metric, data=df_all)
    plt.title(metric)
    plt.xlabel('Categorie')
    plt.ylabel(metric)
plt.tight_layout()
plt.suptitle('Comparaison des métriques par catégorie', y=1.05)
plt.show()
```

```python
# Affichage fusionné : histogramme + barplot comparatif des moyennes pour chaque métrique
import matplotlib.pyplot as plt
import seaborn as sns

# Calcul des moyennes par catégorie pour chaque métrique
df_means = df_all.groupby('Categorie')[metrics].mean().reset_index()

for metric in metrics:
    fig, axes = plt.subplots(1, 2, figsize=(16, 4))

    # Histogramme pour toutes les catégories
    sns.histplot(data=df_all, x=metric, hue='Categorie', element='step', stat='density', common_norm=False, bins=30, palette='tab10', alpha=0.6, ax=axes[0])
    axes[0].set_title(f"Distribution de {metric} par catégorie")
    axes[0].set_xlabel(metric)
    axes[0].set_ylabel('Densité')
    axes[0].legend(title='Categorie')

    # Barplot comparatif des moyennes pour toutes les catégories
    sns.barplot(data=df_means, x='Categorie', y=metric, palette='tab10', ax=axes[1])
    axes[1].set_title(f"Moyenne de {metric} par catégorie")
    axes[1].set_xlabel('Categorie')
    axes[1].set_ylabel(f"Moyenne de {metric}")

    #rajouter les valeurs numériques au-dessus des barres
    for p in axes[1].patches:
        axes[1].annotate(f'{p.get_height():.2f}', 
                         (p.get_x() + p.get_width() / 2., p.get_height()), 
                         ha='center', va='bottom', fontsize=10, color='black')
        
    # rajouter une description des metriques
    

    # Ajustement de la mise en page
    plt.suptitle(f"Analyse de {metric} par catégorie", y=1.05)
    plt.tight_layout()
    plt.show()

```
