```python
# importation des bibliothèques basiques
from pathlib import Path
import sys
import os
from IPython.display import display
import pandas as pd

# Ajout du répertoire racine du projet au PYTHONPATH pour tous les imports
project_root = Path().resolve().parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

# Ensuite importer les modules nécessaires
import src.features.Augmentation.Augmentation as Augmentation
```

```python
project_dir = Path().resolve().parent
print(f"Project directory: {project_dir}")
base_dir = f"{project_dir}/data/raw/COVID-19_Radiography_Dataset/COVID-19_Radiography_Dataset/"
print(f"Base directory: {base_dir}")
classes = ["COVID", "Normal", "Lung_Opacity", "Viral Pneumonia"]
```

```python
import os

for cls in classes:
    images_path = Path(base_dir) / cls / "images"
    print(f"\nTraitement de la classe : {cls}")
    print(f"Chemin des images : {images_path}")
    if not images_path.exists():
        print(f"Le dossier {images_path} n'existe pas !")
        continue
    files = os.listdir(images_path)
    print(f"Nombre de fichiers trouvés : {len(files)}")
    print(f"Exemples : {files[:5]}")
    try:
        Augmentation.process_class(base_dir, cls, n=5)
    except Exception as e:
        print(f"Erreur pour la classe {cls} : {e}")
```
