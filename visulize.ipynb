```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Chemin vers le dossier contenant les images
path_img_covid = "C:\\Users\\serin\\OneDrive\\Bureau\\projet_covid\\COVID-19_Radiography_Dataset\\COVID\\images"  
path_img_normal ="C:\\Users\\serin\\OneDrive\\Bureau\\projet_covid\\COVID-19_Radiography_Dataset\\Normal\\images"
path_img_lung ="C:\\Users\\serin\\OneDrive\\Bureau\\projet_covid\\COVID-19_Radiography_Dataset\\Lung_Opacity\\images"
path_img_pneumonia = "C:\\Users\\serin\\OneDrive\\Bureau\\projet_covid\\COVID-19_Radiography_Dataset\\Viral Pneumonia\\images"

luminosites_covid = []
contrastes_covid  = []

luminosites_normal = []
contrastes_normal  = []

luminosites_pneumo = []
contrastes_pneumo  = []

luminosites_lung = []
contrastes_lung  = []


def calculer_luminosite_contraste(image):
    # Convertit l'image en niveaux de gris
    image_gray = image.convert('L')
    pixels = np.array(image_gray)
    luminosite = np.mean(pixels)
    contraste = np.std(pixels)
    return luminosite, contraste

# Parcours du dossier d'images
for fichier in os.listdir(path_img_covid):
    if fichier.lower().endswith(('.png', '.jpg')):
        chemin_image_covid = os.path.join(path_img_covid, fichier)
        chemin_image_normal = os.path.join(path_img_covid, fichier)
        chemin_image_pneumo = os.path.join(path_img_covid, fichier)
        chemin_image_lung = os.path.join(path_img_covid, fichier)
        try:
            image_covid = Image.open(chemin_image_covid)
            image_normal = Image.open(chemin_image_normal)
            image_pneumo = Image.open(chemin_image_pneumo)
            image_lung = Image.open(chemin_image_lung)
            lum_covid, cont_covid = calculer_luminosite_contraste(image_covid)
            lum_normal, cont_normal = calculer_luminosite_contraste(image_normal)
            lum_lung, cont_lung = calculer_luminosite_contraste(image_pneumo)
            lum_pneumo, cont_pneumo = calculer_luminosite_contraste(image_lung)
            luminosites_covid.append(lum_covid)
            contrastes_covid.append(cont_covid)
            luminosites_normal.append(lum_normal)
            contrastes_normal.append(cont_normal)
            luminosites_pneumo.append(lum_pneumo)
            contrastes_pneumo.append(cont_pneumo)
            luminosites_lung.append(lum_lung)
            contrastes_lung.append(cont_lung)
        except Exception as e: 
            print(f"Erreur avec l'image {fichier} : {e}")

# Affichage des histogrammes
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(luminosites_covid, bins=30, color='yellow', edgecolor='black')
plt.title('Histogramme de la luminosité image Covid')
plt.xlabel('Luminosité')
plt.ylabel("Nombre d'images")

plt.subplot(1, 2, 2)
plt.hist(contrastes_covid, bins=30, color='gray', edgecolor='black')
plt.title('Histogramme du contraste image Covid')
plt.xlabel('Contraste')
plt.ylabel("Nombre d'images")

plt.tight_layout()
plt.show()
```
